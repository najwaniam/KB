## Jumlah label atribut objek

### Subtask:
Tampilkan jumlah nilai unik untuk setiap atribut bertipe objek.


**Reasoning**:
Pilih kolom-kolom dengan tipe data 'object' dan hitung jumlah nilai unik untuk setiap kolom tersebut.



```python
# Pilih kolom-kolom dengan tipe data 'object'
object_cols = df.select_dtypes(include=['object']).columns

# Tampilkan jumlah nilai unik untuk setiap kolom objek
if object_cols.empty:
    print("Tidak ada atribut bertipe objek dalam dataset ini.")
else:
    print("Jumlah nilai unik untuk setiap atribut bertipe objek:")
    for col in object_cols:
        unique_count = df[col].nunique()
        print(f"- {col}: {unique_count}")
```

## Visualisasi data

### Subtask:
Pilih dan buat visualisasi yang sesuai untuk dataset.


**Reasoning**:
Pilih beberapa atribut numerik yang relevan dan buat visualisasi yang sesuai untuk melihat distribusi dan hubungan antar atribut.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atribut numerik yang dipilih untuk visualisasi
selected_numerical_cols = ['temp', 'atemp', 'humidity', 'windspeed', 'hour', 'season', 'year']

# Mengatur gaya visualisasi
sns.set_style("whitegrid")

# Membuat visualisasi untuk setiap atribut yang dipilih
plt.figure(figsize=(15, 10))

# Histogram untuk distribusi
for i, col in enumerate(selected_numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribusi {col.capitalize()}', fontsize=12)
    plt.xlabel(col.capitalize(), fontsize=10)
    plt.ylabel('Frekuensi', fontsize=10)

plt.tight_layout()
plt.show()

# Box plot untuk sebaran dan outlier
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(data=df, y=col)
    plt.title(f'Sebaran {col.capitalize()}', fontsize=12)
    plt.ylabel(col.capitalize(), fontsize=10)

plt.tight_layout()
plt.show()

# Scatter plot untuk hubungan antara temp dan atemp
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='temp', y='atemp')
plt.title('Hubungan antara Suhu dan Suhu Perasaan', fontsize=14)
plt.xlabel('Suhu (°C)', fontsize=12)
plt.ylabel('Suhu Perasaan (°C)', fontsize=12)
plt.show()

# Bar plot untuk rata-rata temp per season
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='season', y='temp')
plt.title('Rata-rata Suhu per Musim', fontsize=14)
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Rata-rata Suhu (°C)', fontsize=12)
plt.xticks(ticks=[0, 1, 2, 3], labels=['Semi', 'Panas', 'Gugur', 'Dingin'])
plt.show()
```

## Korelasi menggunakan heatmap

### Subtask:
Hitung dan visualisasikan matriks korelasi menggunakan heatmap.


**Reasoning**:
Hitung matriks korelasi untuk DataFrame `df` dan visualisasikan menggunakan heatmap dengan anotasi dan ukuran gambar yang sesuai, serta tambahkan judul.



```python
# Hitung matriks korelasi untuk DataFrame df
correlation_matrix = df.corr(numeric_only=True)

# Atur ukuran gambar heatmap
plt.figure(figsize=(12, 10))

# Buat visualisasi heatmap dari matriks korelasi
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Berikan judul pada heatmap
plt.title('Matriks Korelasi Atribut Numerik', fontsize=16)

# Tampilkan plot
plt.show()
```
